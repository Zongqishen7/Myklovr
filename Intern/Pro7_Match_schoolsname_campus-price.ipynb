```python
def drop_pun(str):
    str = re.sub(r'[^\w\s]','',str)
    str = re.sub(' ','',str)
    str = re.sub('university','1',str)
    str = re.sub('college','2',str)
    str = re.sub('the','3',str)
    return str

def to_lower(item):
    return item.lower()

#
def stringCleaning(str):
    if str is None:
        return None
    try:
        str = to_lower(str)
        str = drop_pun(str)
        
        return str
    except:
        return None

stringCleaning('ab.`college')
```
